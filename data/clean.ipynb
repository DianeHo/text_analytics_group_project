```python
import pandas as pd
import numpy as np

import tba3102

# Load the dataset
df = pd.read_csv('tariff_malaysia.csv')
df.head()

tba3102.data_quality_report(df)
```

```python
df['Comment Body'].replace(r'^(\s)+$', np.nan, regex=True, inplace=True)
df = df.dropna(subset=['Comment Body']).reset_index(drop=True)
tba3102.data_quality_report(df)
```

```python
# Normalize the corpus, excluding specific terms from cleaning
tba3102.normalize_corpus(df, 'Comment Body', 'Cleaned_Comment')

# Generate a data quality report after cleaning
tba3102.data_quality_report(df)
```

```python
df['Cleaned_Comment'].replace(r'^(\s)+$', np.nan, regex=True, inplace=True)
df = df.dropna(subset=['Cleaned_Comment']).reset_index(drop=True)
tba3102.data_quality_report(df)
```

```python
df.to_csv('cleaned-tariff_malaysia.csv', index=False)
```
