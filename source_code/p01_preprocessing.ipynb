Part 1 Text Preproccessing

```python
#import necessary libraries
import pandas as pd
import re
import tba3102
import nltk
import numpy as np

import spacy

from datetime import datetime
```

```python
#load dataset
filepath = '../data/iphone.xlsx'
df = pd.read_excel(filepath)

display(df.head(10))
print('Number of rows:', df.shape[0])
```

```python
df.isna().sum()
```

```python
#remove column links
df.drop('Links', axis=1, inplace=True)
df.reset_index(drop=True, inplace=True)

print(df.isna().sum())
print('Number of rows:', df.shape[0])
```

```python
#remove empty rows
df.dropna(inplace=True)
df.reset_index(drop=True, inplace=True)

print(df.isna().sum())
print('Number of rows:', df.shape[0])
```

```python
#text-processing (correct spelling, lemmatization, removal of stop words)
tba3102.set_default_pandas_options()

print('Text processing started at {}'.format(datetime.now()))
```

```python
stopword_exclusion_list = ['no', 'not']
tba3102.exclude_stopwords(stopword_exclusion_list)


tba3102.normalize_corpus(df, 'Message', 'Cleaned_Message')

df.head(10)
```

```python
#check for empty strings and replace with NaN
df = df.replace(r'^(\s?)+$', np.nan, regex=True)
print(df.isna().sum())
print('Number of rows:', df.shape[0])
```

```python
df = df.dropna().reset_index(drop=True)
print(df.isna().sum())
print('Number of rows:', df.shape[0])
```

```python
df.to_csv('cleaned_message.csv', index=False)
```
